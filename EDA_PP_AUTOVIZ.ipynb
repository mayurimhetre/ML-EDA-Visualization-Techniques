```python
import sweetviz as sv

df = pd.read_csv('car data.csv')
df
sweet_report = sv.analyze([df,"train"],'Selling_Price')
sweet_report.show_html('sweet_report.html')
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Selling_Price',axis=1)
y = df['Selling_Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
df1 = pd.DataFrame(y_train)
```

```python
df_new = pd.concat([X_train,df1],axis=1)
df_new.head()
```

```python
X_test.head()
```

```python
################## SWEETVIZ COMPARE ####################
sweet_report_cmp = sv.compare([df_new,"train"],[X_test,"test"],'Selling_Price')
sweet_report_cmp.show_html('sweet_report_cmp.html')
```
